```python
# agrupamento com o K-means
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
dados = make_blobs(n_samples=200, n_features=2, centers=5, cluster_std=1, random_state=150)
plt.scatter(dados[0][:,0], dados[0][:,1], c=dados[1],cmap='rainbow')
```

```python
wcss = []
for i in range (1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 150)
    kmeans.fit(dados[0])
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11), wcss, color='k')
plt.title('A Curva do Cotovelo do Algoritmo K-means')
plt.xlabel('Número de Clusters (K)')
plt.ylabel('Soma dos Quadrados do Cluster (wcss)')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++')
kmeans.fit(dados[0])
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10,6))
ax1.set_title('K-means')
ax1.scatter(dados[0][:,0],dados[0][:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title('Original')
ax2.scatter(dados[0][:,0],dados[0][:,1],c=dados[1],cmap='rainbow')
```
